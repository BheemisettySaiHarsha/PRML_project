```python
u_id_1 = sorted(set(map(int, tags.values[:, 0])))

u_id_2=sorted(set(map(int, ratings.values[:, 0])))

u_id_1.extend(u_id_2)     # user id's
u_id_1=sorted(list(set(u_id_1)))
print(u_id_1)
```

```python
df=pd.DataFrame(Data)
df.head()
```

```python
print(df.columns)
```

```python
movie_dataset=pd.DataFrame({})
movie_dataset['movieId'] =df['Movie ID']
movie_dataset['movie title'] =df['Movie Title']
movie_dataset.head()
```

```python
merged_dataset = pd.merge(ratings, movie_dataset, how='inner', on='movieId')
merged_dataset.head()
```

```python
refined_dataset = merged_dataset.groupby(by=['userId','movie title'], as_index=False).agg({"rating":"mean"})
refined_dataset.head()
```

```python
user_to_movie_df = refined_dataset.pivot(
    index='userId',
     columns='movie title',
      values='rating').fillna(0)

user_to_movie_df.head()
```
