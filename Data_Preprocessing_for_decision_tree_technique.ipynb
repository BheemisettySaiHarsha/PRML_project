```python
import pandas as pd
```

```python
ratings_df = pd.read_csv('https://raw.githubusercontent.com/BheemisettySaiHarsha/PRML_project/main/PRML%20DATASET/ratings.csv')
movies_df = pd.read_csv('https://raw.githubusercontent.com/BheemisettySaiHarsha/PRML_project/main/PRML%20DATASET/movies.csv')
```

```python
combined_df = pd.merge(movies_df, ratings_df, on='movieId')
combined_df = combined_df.drop(['genres', 'timestamp'], axis=1)
```

```python
combined_df
```

```python
combined_df['avg_rating'] = combined_df.groupby('movieId')['rating'].transform('mean')
```

```python
combined_df
```

```python
filtered_df = combined_df[combined_df.groupby('movieId')['rating'].transform('count') > 50]
```

```python
filtered_df
```

```python
rating_matrix = filtered_df.pivot(index='userId', columns='movieId', values='rating').fillna(0)
```

```python
rating_matrix
```

```python
# First compute user rating means
user_means = rating_matrix.mean(axis=1)

# Subtract user_means from rating_matrix, replacing NaN with 0
normalized_matrix = rating_matrix.sub(user_means, axis=0).fillna(0)
```
